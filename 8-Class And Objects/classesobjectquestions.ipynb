```python
#Question 5+6
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age=age
class Employee(Person):
    def __init__(self, name, age,employee_id):
        super().__init__(name, age)
        self.employee_id = employee_id

    def __str__(self):
        return f"Employee(Name: {self.name}, Age: {self.age}, Employee ID: {self.employee_id})"


p1 = Employee("yassmine",24,1)
print("the name of employee is :",p1.name)
print("the age of employee is :",p1.age)
print("the id of employee is :",p1.employee_id)
print(p1)
```

```python
#Question 9
import math

class MathOperations:
    @staticmethod
    def sqrt(x):
        return math.sqrt(x)
print(MathOperations.sqrt(16))
```

```python
#Question 10
class Rectangle:
    def __init__(self, length, width):
        self.__length = length
        self.__width = width

    @property
    def length(self):
        return self.__length

    @length.setter
    def length(self, length):
        self.__length = length

    @property
    def width(self):
        return self.__width

    @width.setter
    def width(self, width):
        self.__width = width


rect = Rectangle(10, 5)
print(rect.length, rect.width)  
rect.length = 15
rect.width = 7
print(rect.length, rect.width)  
```
